```python
from __future__ import print_function
import pylab as pl 
import numpy as np

#import seaborn

from scipy.optimize import curve_fit, minimize
%pylab inline
```

```python
#distributions = ['pois', 'gaus', 'chisq', 'cauchy', 'lnorm', 'binomial']
distributions = ['chisq']
```

```python
my_mean = 100
df = my_mean
```

```python
md = {}
md['chisq'] = np.random.chisquare(df, size = 100)
pl.hist(md['chisq'], bins = 30)
pl.ylabel('N')
pl.xlabel('x')
```

```python
print ("Chisq mean: %.2f, Standard deviation: %.2f"%(md['chisq'].mean(),\
md['chisq'].std()))
```

```python
my_size = (2000 / (np.array(range(1, 100)))).astype(int)
print(my_size, my_size.shape)
```

```python
md['chisq'] = {}

for i in my_size:
    md['chisq'][i] = np.random.chisquare(df, size = i)
    
md['chisq']['means'] = {}
axchisq_mu_i = pl.figure(figsize=(10,6)).add_subplot(111)

for ii in md['chisq']:
    if not type(ii) == str:
        md['chisq']['means'][ii] = md['chisq'][ii].mean()
        axchisq_mu_i.plot(ii, md['chisq']['means'][ii], 'o')
        axchisq_mu_i.set_xlabel('sample size', fontsize=20)
        axchisq_mu_i.set_ylabel('sample size', fontsize=20)
        axchisq_mu_i.set_title('Chi squared', fontsize=20)
        axchisq_mu_i.plot([min(my_size), max(my_size)], [df, df], 'k')
```

```python
allmeans = list(md['chisq']['means'].values())

pl.figure(figsize=(5, 5))
pl.hist(allmeans, bins=30)
pl.xlabel('sample mean', fontsize=18)
pl.ylabel('N', fontsize=18)
```
